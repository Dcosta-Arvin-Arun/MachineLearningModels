```python
# step 1: import library
import pandas as pd
```

```python
# step 2: import data
diabetes = pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Diabetes.csv")
```

```python
# step 3: define target (y) and features (x)
y = diabetes['diabetes']
X = diabetes[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
# step 4: train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.75, random_state=2529)
```

```python
# step 5: Select model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=500)
```

```python
# step 6: train model (fit model)
model.fit(X_train, y_train)
```

```python
# step 7: prediction
y_pred = model.predict(X_test)
```

```python
# step 8: accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred) * 100
print(accuracy)
```
