**House Price prediction**

```python
# step 1: import library
import pandas as pd
```

```python
# step 2: import data
house = pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Boston.csv")
```

```python
# step 3: define target (y) and features (x)
y = house["MEDV"]
X = house[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
X.shape
```

```python
y.shape
```

```python
# step 4: train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.75, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train
```

```python
# step 5: Select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# step 6: train model (fit model)
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# step 7: prediction
y_pred = model.predict(X_test)
print(y_pred)
```

```python
# step 8: accuracy
from sklearn.metrics import mean_absolute_percentage_error
accuracy = 100 - mean_absolute_percentage_error(y_test, y_pred) * 100
print(accuracy)
```
